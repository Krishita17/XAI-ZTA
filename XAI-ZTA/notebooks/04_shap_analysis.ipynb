# 04 - SHAP Analysis

Deep SHAP explanation analysis for authentication decisions.

## Analysis Plan
- TreeExplainer for Random Forest and XGBoost
- Waterfall plots for individual decisions
- Force plots for push/pull visualization
- Summary plot for global feature importance
- Dependency plots for feature interactions

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import shap
from src.xai.shap_explainer import SHAPExplainer
from src.models.train import prepare_data
from src.models.random_forest import build_random_forest, train_random_forest
```

```python
# Prepare data and train model
X_train, X_test, y_train, y_test, feature_names = prepare_data()
rf = build_random_forest()
result = train_random_forest(rf, X_train, y_train)
model = result['model']
```

```python
# Initialize SHAP explainer
explainer = SHAPExplainer(model, model_type='tree', feature_names=feature_names)

# Explain a single decision
explanation = explainer.explain_single(X_test[0])
print('Top contributing features:')
for f in explanation['top_features'][:5]:
    print(f"  {f['feature']}: {f['shap_value']:+.4f}")
```

```python
# Global feature importance
importance = explainer.get_global_importance(X_test[:100])
print('\nGlobal Feature Importance:')
for name, score in importance.items():
    print(f'  {name}: {score:.4f}')
```
