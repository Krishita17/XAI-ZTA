LIME explanation experiments for authentication decisions.

## Objectives
- Generate LIME explanations for individual decisions
- Compare LIME across different model types
- Analyze explanation consistency
- Evaluate explanation quality

```python
import sys
sys.path.insert(0, '..')

import numpy as np
from src.xai.lime_explainer import LIMEExplainer
from src.models.train import prepare_data
from src.models.random_forest import build_random_forest, train_random_forest
```

```python
# Prepare data and train model
X_train, X_test, y_train, y_test, feature_names = prepare_data()
rf = build_random_forest()
result = train_random_forest(rf, X_train, y_train)
model = result['model']
```

```python
# Initialize LIME explainer
lime_exp = LIMEExplainer(X_train, feature_names=feature_names)

# Explain a single decision
explanation = lime_exp.explain_single(X_test[0], model.predict_proba)
print(explanation['text_summary'])
```
