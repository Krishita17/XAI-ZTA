# 01 - Data Exploration

Exploratory Data Analysis on authentication datasets for XAI-ZTA.

## Objectives
- Load and inspect synthetic authentication event data
- Analyze feature distributions
- Check class balance (ALLOW vs DENY)
- Identify correlations between features
- Visualize key patterns

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from src.data.loader import load_synthetic_data, get_data_summary

# Set plotting style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Load synthetic authentication data
df = load_synthetic_data('data/synthetic/generated_auth_logs.csv')
print(f'Dataset shape: {df.shape}')
print(f'\nColumn types:\n{df.dtypes}')
df.head()
```

```python
# Class distribution
print('Decision distribution:')
print(df['decision'].value_counts())
print(f'\nClass balance: {df["decision"].value_counts(normalize=True).round(3).to_dict()}')

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
df['decision'].value_counts().plot(kind='bar', ax=ax[0], color=['#28a745', '#dc3545'])
ax[0].set_title('Decision Distribution')
ax[0].set_ylabel('Count')

# Trust score distribution
# TODO: Add device_trust_score histogram by decision
plt.tight_layout()
plt.show()
```

```python
# Feature correlation matrix
numeric_cols = df.select_dtypes(include=[np.number]).columns
corr = df[numeric_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap='RdBu_r', center=0, fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```
