Compare SHAP vs LIME vs Anchor explanation methods.

## Evaluation Metrics (Novel Contribution)
- **Faithfulness**: Does removing top features degrade performance?
- **Stability**: Same input → same explanation across runs?
- **Sparsity**: How many features needed to explain?
- **Comprehensibility**: Readable by non-ML analysts?
- **Latency**: Time to generate (must be < 500ms)

```python
import sys
sys.path.insert(0, '..')

import numpy as np
from src.xai.xai_evaluator import XAIEvaluator
from src.models.train import prepare_data
from src.models.random_forest import build_random_forest, train_random_forest
```

```python
# Prepare data and train model
X_train, X_test, y_train, y_test, feature_names = prepare_data()
rf = build_random_forest()
result = train_random_forest(rf, X_train, y_train)
model = result['model']
```

```python
# Initialize evaluator
evaluator = XAIEvaluator(model, feature_names=feature_names)

# Evaluate faithfulness using feature importances
faithfulness = evaluator.evaluate_faithfulness(
    X_test, y_test, model.feature_importances_
)
print(f"Faithfulness score: {faithfulness['faithfulness_score']}")
```
